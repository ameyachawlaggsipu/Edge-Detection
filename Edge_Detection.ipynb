## Edge Detection Using Canny

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("download (1).jfif", 0)
img2 = cv2.imread("download.png", 0)
```

```python
canny = cv2.Canny(img1, 100, 200)
```

```python
titles = ["image", "canny"]
```

```python
images = [img1, canny]
```

```python
for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
plt.show()
```

```python
canny2 = cv2.Canny(img2, 100, 200)
```

```python
titles = ["image2", "canny2"]
```

```python
images = [img2, canny2]
```

```python
for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
plt.show()
```
